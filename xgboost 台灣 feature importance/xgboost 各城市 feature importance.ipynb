```python
# 
from sklearn.model_selection import train_test_split
import xgboost as xgb
from xgboost import XGBRegressor
from sklearn.feature_selection import SelectFromModel
```

```python
x10 = New_Taipei.iloc[:10, 0:5]
y10 = New_Taipei.iloc[:10, 5]
x20 = Taipei.iloc[:10, 0:5]
y20 = Taipei.iloc[:10, 5]
x30 = Taoyuan.iloc[:10, 0:5]
y30 = Taoyuan.iloc[:10, 5]
x40 = Taichung.iloc[:10, 0:5]
y40 = Taichung.iloc[:10, 5]
x50 = Tainan.iloc[:10, 0:5]
y50 = Tainan.iloc[:10, 5]
x60 = Kaohsiung.iloc[:10, 0:5]
y60 = Kaohsiung.iloc[:10, 5]

x11 = New_Taipei.iloc[:, 0:5]
y11 = New_Taipei.iloc[:, 5]
x21 = Taipei.iloc[:, 0:5]
y21 = Taipei.iloc[:, 5]
x31 = Taoyuan.iloc[:, 0:5]
y31 = Taoyuan.iloc[:, 5]
x41 = Taichung.iloc[:, 0:5]
y41 = Taichung.iloc[:, 5]
x51 = Tainan.iloc[:, 0:5]
y51 = Tainan.iloc[:, 5]
x61 = Kaohsiung.iloc[:, 0:5]
y61 = Kaohsiung.iloc[:, 5]
```

```python
fea10 = x10.copy()
fea20 = x20.copy()
fea30 = x30.copy()
fea40 = x40.copy()
fea50 = x50.copy()
fea60 = x60.copy()
```

```python
x10.info()
```

```python
x10 = x10.values
x20 = x20.values
x30 = x30.values
x40 = x40.values
x50 = x50.values
x60 = x60.values

x11 = x11.values
x21 = x21.values
x31 = x31.values
x41 = x41.values
x51 = x51.values
x61 = x61.values
```

```python
y10 = y10.values
y20 = y20.values
y30 = y30.values
y40 = y40.values
y50 = y50.values
y60 = y60.values

y11 = y11.values
y21 = y21.values
y31 = y31.values
y41 = y41.values
y51 = y51.values
y61 = y61.values
```

### (2009~2018)

```python
xgbr = XGBRegressor(n_estimators=100 , max_depth=2 , learning_rate=0.1, subsample=1, colsample_bynode=0.8, reg_lambda=1e-05)
xgbr.fit(x10, y10)

# get importance
importance = xgbr.feature_importances_

# summarize feature importance
for i,v in enumerate(importance):
    print('Feature: %0d, Score: %.5f' % (i+1,v))

# plot feature importance
plt.bar([x for x in range(len(importance))], importance)
plt.show()
plt.figure(figsize=(50, 50))
importances = pd.DataFrame({'feature':fea10.columns,'importance':np.round(xgbr.feature_importances_,3)}) 
importances = importances.sort_values('importance',ascending=False).set_index('feature')  
importances.plot.bar()
```
