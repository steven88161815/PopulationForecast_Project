```python
x0 = df.iloc[:60, 0:5]
y0 = df.iloc[:60, 5]


x1 = df.iloc[60:, 0:5]
y1 = df.iloc[60:, 5]
```

```python
x0 = x0.values
x1 = x1.values
```

```python
y0 = y0.values
y1 = y1.values
```

### (2009~2018)

```python
xgbr = XGBRegressor(n_estimators=100 , max_depth=3 , learning_rate=0.1, subsample=1, colsample_bynode=0.8, reg_lambda=1e-05)
xgbr.fit(x0, y0)

# get importance
importance = xgbr.feature_importances_

# summarize feature importance
for i,v in enumerate(importance):
    print('Feature: %0d, Score: %.5f' % (i+1,v))

# plot feature importance
plt.bar([x for x in range(len(importance))], importance)
plt.show()
plt.figure(figsize=(50, 50))
importances = pd.DataFrame({'feature':fea0.columns,'importance':np.round(xgbr.feature_importances_,3)}) 
importances = importances.sort_values('importance',ascending=False).set_index('feature')  
importances.plot.bar()
```

### (2009~2025)

```python
xgbr = XGBRegressor(n_estimators=100 , max_depth=2 , learning_rate=0.1, subsample=1, colsample_bynode=0.8, reg_lambda=1e-05)
xgbr.fit(x1, y1)

# get importance
importance = xgbr.feature_importances_

# summarize feature importance
for i,v in enumerate(importance):
    print('Feature: %0d, Score: %.5f' % (i+1,v))

# plot feature importance
plt.bar([x for x in range(len(importance))], importance)
plt.show()
plt.figure(figsize=(50, 50))
importances = pd.DataFrame({'feature':fea0.columns,'importance':np.round(xgbr.feature_importances_,3)}) 
importances = importances.sort_values('importance',ascending=False).set_index('feature')  
importances.plot.bar()
```
