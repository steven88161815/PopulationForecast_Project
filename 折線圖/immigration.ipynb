### input為n維

````python
d = int(input('輸入的維度：'))
x_train = []
y_train = []


for i in range(10-d):
    x_train.append([])
    y_train.append([])
    
for j in range(10-d):
    x_train[j] = (df1.iloc[:, j:d+j].values)
    y_train[j] = (df1.iloc[:, d+j].values)
````

````python
x_train = np.array(x_train)
y_train = np.array(y_train)
````

````python
x_train
````

````python
x_train.shape
````

````python
y_train
````

````python
y_train.shape
````

````python
x_train = x_train.reshape(-1, d)
y_train = y_train.ravel()
````

````python
x_train
````

````python
y_train
````

````python
from xgboost import XGBRegressor
xgbr = XGBRegressor(n_estimators=100 , max_depth=3 , learning_rate=0.1, subsample=1, colsample_bynode=0.8, reg_lambda=1e-05)
xgbr.fit(x_train, y_train)
````

````python
# 計算各個城市的MAPE
box1 = []
for i in range(len(y_train)):
    MAPE = abs((xgbr.predict(x_train)[i]-y_train[i])/y_train[i])
    box1.append(MAPE)

box2 = []
for j in range(6):
    num = 0
    for k in range(j, len(y_train), 6):
        num += box1[k]
    num = num/(len(y_train)/6)
    box2.append(num)

for a in range(6):
    print('MAPE為', box2[a])
````

````python
num = 0
for i in range(6):
    num += box2[i]
num /= 6
print('平均MAPE為', num)
````
